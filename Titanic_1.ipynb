```python
train.head()
```

```python
train['title']=train['Name'].apply(lambda X: X.split(',')[1].split('.')[0].strip())
train.head()
```

```python
train.title.value_counts()
```

```python
dictionary={'Mr':'Mr','Miss':'Miss', 'Mrs':'Mrs', 'Master':'Master','Dr':'Dr','Rev':'Rev','Mlle':'Miss','Col':'Mr',
           'Major':'Mr','Lady':'Miss','Jonkheer':'Master','Don':'Mr','Ms':'Miss','Mme':'Mrs','Capt':'Mr','the Countess':'Mrs',
           'Sir':'Mr'} 
train['title']=train['title'].apply(lambda X: dictionary[X])
```

```python
train.title.value_counts()
```

```python
train['Sex']=train['Sex'].apply(lambda X: 1 if X=='female' else 0)
```

```python
train.Sex.value_counts()
```

```python
pd.crosstab(train.title, train.Survived)
```

```python
dictionary_1={'Dr':4,'Master':3,'Miss':1,'Mr':5,'Mrs':2,'Rev':6}
train['title']=train['title'].apply(lambda X: dictionary_1[X])
```

```python
train.title.value_counts()
```

```python
train.head()
```

```python
train.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
train.head()
```

```python
pd.crosstab(train.Embarked, train.Survived)
```

```python
help(pd.get_dummies)
```

```python
train=pd.get_dummies(train,prefix='EMB',columns=['Embarked'],drop_first=True)
train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
#train_df, test_df=train_test_split(train,)
help(train_test_split)
```

```python
train_df, test_df=train_test_split(train,test_size=0.2,random_state=10, stratify=train.Survived)
```

```python
tree=DecisionTreeClassifier(random_state=10)
dictionari_2={'max_depth': [4,5,6,7,8], 'min_samples_split':[0.1,0.01]}
grid=GridSearchCV(tree,param_grid=dictionari_2,cv=5)
grid.fit(train.drop('Survived',1),train.Survived)
print grid.best_params_
```

```python
dictionari_2={'max_depth': [2,3,4], 'min_samples_split':np.linspace(0.01,0.05,5)}
grid=GridSearchCV(tree,param_grid=dictionari_2,cv=5)
grid.fit(train.drop('Survived',1),train.Survived)
print grid.best_params_
```

```python
grid.cv_results_
```

```python
tree=DecisionTreeClassifier(max_depth=3, min_samples_split=0.01,random_state=10)
dictionari_2={'min_samples_leaf':[0.01,0.001,0.0001]}
grid=GridSearchCV(tree,param_grid=dictionari_2,cv=5)
grid.fit(train.drop('Survived',1),train.Survived)
print grid.best_params_
```

```python
tree=DecisionTreeClassifier(max_depth=3, min_samples_split=0.01,min_samples_leaf=0.001,random_state=10)
tree.fit(train_df.drop('Survived',1),train_df.Survived)
predict=tree.predict(test_df.drop('Survived',1))
print accuracy_score(test_df.Survived,predict)
print roc_auc_score(test_df.Survived,predict)
```
